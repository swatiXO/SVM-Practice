Importing the required things for SVM


```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = digits.data
y = digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Training with Both kernels

```python
classifier = SVC(kernel = 'rbf')
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
modelLinear = SVC (kernel='linear')
```

```python
modelLinear.fit(X_train,y_train)
```

```python
modelLinear.score(X_test,y_test)
```

RBF kernel has better score than Linear kernel
Lets try to improve the rbf kernel

```python
model_C = SVC(C=1,kernel = "rbf")
```

```python
model_C.fit(X_train, y_train)
```

```python
model_C.score(X_test, y_test)
```

RBF kernel gives the highest

```python
model_D = SVC (kernel = 'rbf',gamma = 0.10)
```

```python
model_D.fit(X_train,y_train)
```

```python
model_D.score(X_test, y_test)
```

Tuning the gamma has ruined it for us, lets try tuning the C parameter instead

```python
model_F = SVC(kernel = 'rbf',C=10)
```

```python
model_F.fit(X_train, y_train)
```

```python
model_F.score(X_train, y_train)
```

Our score has improved from 0.98 to 1, which means it might have been over fit.

```python
model_X = SVC(kernel = 'rbf',C=5)
```

```python
model_X.fit(X_train, y_train)
```

```python
model_X.score(X_train, y_train)
```

this is also pretty close to 1. We have tried our best at tuning

Lets use grid Search for a better estimate

```python
from sklearn.model_selection import GridSearchCV
```

# Setting range of values for C and Gamma params

```python

param_grid = {'C': [0.1, 1, 10,20,30, 100], 'gamma': [0.001, 0.01, 0.1, 1]}

```

# Creating an SVM classifier with an 'rbf' kernel since it worked best for us

```python


clf = SVC(kernel='rbf')
```


# Performing grid search

```python

grid_search = GridSearchCV(clf, param_grid)
grid_search.fit(X_train, y_train)
```

# Get the best parameters and model

```python


best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

```

# Make predictions using the best model

```python

y_pred_best = best_model.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)

print(f"Best Parameters: {best_params}")
print(f"Accuracy with Best Parameters: {accuracy_best * 100:.2f}%")
```

# According to gridsearch, our best params are c= 1 , and gamma = 0.001
